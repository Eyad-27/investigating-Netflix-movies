```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Subset the DataFrame to keep only movies
movies_df = netflix_df[netflix_df['type'] == 'Movie']

# Filter the DataFrame for movies released between 1990 and 1999
movies_90s_df = movies_df[(movies_df['release_year'] >= 1990) & (movies_df['release_year'] <= 1999)]

# Select the title and release_year columns
movies_90s_titles_years = movies_90s_df[['title', 'release_year']]

# Display the resulting DataFrame
movies_90s_titles_years
```

```python
# Plotting the distribution of movie durations for movies released in the 1990s
plt.figure(figsize=(10, 6))
plt.hist(movies_90s_df['duration'], bins=20, edgecolor='black')
plt.title('Distribution of Movie Durations (1990s)')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.show()
```

```python
# Finding the most frequent movie duration in the 1990s
most_frequent_duration = movies_90s_df['duration'].mode()[0]

# Saving the result as an integer called duration
duration = int(most_frequent_duration)

# Display the result
duration
```

```python
# Subset the DataFrame to keep only movies with the genre 'Action' and released in the 1990s
action_movies_90s_df = movies_df[(movies_df['genre'] == 'Action') & 
                                 (movies_df['release_year'] >= 1990) & 
                                 (movies_df['release_year'] <= 1999)]

# Initialize a counter for action movies with duration less than 90 minutes
short_movie_count = 0

# Iterate through the DataFrame and count action movies with duration less than 90 minutes
for index, row in action_movies_90s_df.iterrows():
    if row['duration'] < 90:
        short_movie_count += 1

# Display the count of short action movies
short_movie_count
```
